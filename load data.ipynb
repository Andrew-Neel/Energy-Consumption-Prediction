```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("zurich_electricity_consumption.csv")
```

```python
print("First 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())
```

```python
print("\nShape of dataset:")
print(df.shape)

print("\nColumn names:")
print(df.columns.tolist())

print("\nData types:")
print(df.dtypes)
```

```python
# Replace 'DateTime' with actual datetime column name if different
df["Timestamp"] = pd.to_datetime(df["Timestamp"], errors="coerce")

# Set datetime as index (recommended for time series)
df = df.set_index("Timestamp").sort_index()

print("\nDatetime index preview:")
print(df.index)
```

```python
print("\nMissing values by column:")
print(df.isnull().sum())
```

```python
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
df[['Value_NE5','Value_NE7']].corr()
df[['Value_NE5','Value_NE7']].describe()
```

```python
print("\nFirst 10 timestamp differences:")
print(df.index.to_series().diff().value_counts().head(10))

# Expected for 15-minute data:
# 0 days 00:15:00
```

```python
print("\nNumerical summary:")
print(df.describe())
```

```python
Q1 = df['Value_NE5'].quantile(0.25)
Q3 = df['Value_NE5'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers= ((df['Value_NE5'] < lower) | (df['Value_NE5'] > upper)).astype(int)

print("Outliers detected:", outliers.sum())
```

```python
full_range = pd.date_range(
    start=df.index.min(),
    end=df.index.max(),
    freq="15min"
)

missing_timestamps = full_range.difference(df.index)

print("\nNumber of missing timestamps:", len(missing_timestamps))

if len(missing_timestamps) > 0:
    print("\nFirst 10 missing timestamps:")
    print(missing_timestamps[:10])
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    negatives = (df[col] < 0).sum()
    if negatives > 0:
        print(f"{col}: {negatives} negative values")
```
